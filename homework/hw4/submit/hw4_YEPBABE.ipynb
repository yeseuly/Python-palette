[문제1] 커피 데이터의 대륙별 레코드 비율이 어떻게 되나요? 대륙별 커피원두의 색상 분포를 확인해주세요! (nan 제외)
- ['Continent.of.Origin'] = ['Africa', 'North America', 'South America', 'Asia', 'Oceania', nan]
- ['Color'] = ['Green', nan, 'Bluish-Green', 'Blue-Green']

```python
Africa = data[data['Continent.of.Origin'] == 'Africa']
North_America = data[data['Continent.of.Origin'] == 'North America']
South_America = data[data['Continent.of.Origin'] == 'South America']
Asia = data[data['Continent.of.Origin'] == 'Asia']
Oceania = data[data['Continent.of.Origin'] == 'Oceania']
```

```python
print("아프리카의 커피생산 비율은", len(Africa)/len(data)*100,'%,\n' 
      "북미의 커피생산 비율은", len(North_America)/len(data)*100,'%,\n'
      "남미의 커피생산 비율은", len(South_America)/len(data)*100,'%,\n'
      "아시아의 커피생산 비율은", len(Asia)/len(data)*100,'%,\n'
      "오세아니아의 커피생산 비율은", len(Oceania)/len(data)*100,'%,\n')
```

```python
### 대륙별 커피원두의 색상분포(비율)
### ['Color'] = ['Green', nan, 'Bluish-Green', 'Blue-Green']
Green_Af = Africa[Africa['Color']=='Green']
BluishGreen_Af = Africa[Africa['Color']=='Bluish-Green']
BlueGreen_Af = Africa[Africa['Color']=='Blue-Green']

print('아프리카대륙 커피중 에서 Green색상은', len(Green_Af)/len(Africa)*100,'%,\n'
      '아프리카대륙 커피중 에서 Bluish-Green색상은', len(BluishGreen_Af)/len(Africa)*100,'%,\n'
      '아프리카대륙 커피중 에서 Blue-Green색상은', len(BlueGreen_Af)/len(Africa)*100,'%,\n')
print('다른대륙은 노가다라 안했음')
```

[문제2] 커피원두 대륙별 + 색상별로 산도가 얼마나 다른가요? 산도를 확인할 수 있는 그래프를 그려주세요. (data['Acidity]>6)
- data_acidity = data_*[data_*['Acidity']>6] (data_*: 특정대륙데이터)
- plt.hist(data_acidity['Acidity'], bins=100, edgecolor='black')

```python
##산도 6이상에 대해서만
아프_산도 = Africa[Africa['Acidity'] > 6]
아시아_산도 = Asia[Asia['Acidity'] > 6]
북미_산도 = North_America[North_America['Acidity'] > 6]
남미_산도 = South_America[South_America['Acidity'] > 6]
오세_산도 = Oceania[Oceania['Acidity'] > 6]

plt.figure(figsize=(13,8))
plt.subplot(2,3,1)
a= plt.hist(아프_산도['Acidity'], bins=100, edgecolor='black')
plt.title("Africa Coffee Acidity")
plt.subplot(2,3,2)
a= plt.hist(아시아_산도['Acidity'], bins=100, edgecolor='black')
plt.title("Asia Coffee Acidity")
plt.subplot(2,3,3)
a= plt.hist(북미_산도['Acidity'], bins=100, edgecolor='black')
plt.title("N America Coffee Acidity")
plt.subplot(2,3,4)
a= plt.hist(남미_산도['Acidity'], bins=100, edgecolor='black')
plt.title("S America Coffee Acidity")
plt.subplot(2,3,5)
a= plt.hist(오세_산도['Acidity'], bins=100, edgecolor='black')
plt.title("Oce Coffee Acidity")
```
