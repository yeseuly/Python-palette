[문제1] 커피 데이터의 대륙별 레코드 비율이 어떻게 되나요? 대륙별 커피원두의 색상 분포를 확인해주세요! (nan 제외)
- ['Continent.of.Origin'] = ['Africa', 'North America', 'South America', 'Asia', 'Oceania', nan]
- ['Color'] = ['Green', nan, 'Bluish-Green', 'Blue-Green']

```python
# 대륙별 분류
africa = data[data['Continent.of.Origin'] == 'Africa']
north_america = data[data['Continent.of.Origin'] == 'North America']
south_america = data[data['Continent.of.Origin'] == 'South America']
asia = data[data['Continent.of.Origin'] == 'Asia']
oceania = data[data['Continent.of.Origin'] == 'Oceania']


# 대륙별 레코드 비율
per_africa = len(africa)*100/len(data)
per_north_america = len(north_america)*100/len(data)
per_south_america = len(south_america)*100/len(data)
per_asia = len(asia)*100/len(data)
per_oceania = len(oceania)*100/len(data)

print("Africa 원두는", round(per_africa, 2), "% 입니다.")
print("North America 원두는", round(per_north_america, 2), "% 입니다.")
print("South America 원두는", round(per_south_america, 2), "% 입니다.")
print("Asia 원두는", round(per_asia, 2), "% 입니다.")
print("Oceania 원두는", round(per_oceania, 2), "% 입니다.")


# 대륙별 커피원두의 색상 분포
per_green = len(data[data['Color'] == 'Green'])*100/len(data)
per_bluish_green = len(data[data['Color'] == 'Bluish-Green'])*100/len(data)
per_blue_green = len(data[data['Color'] == 'Blue-Green'])*100/len(data)

print("Green 원두는", round(per_green, 2), "% 입니다.")
print("Bluish-Green 원두는", round(per_bluish_green, 2), "% 입니다.")
print("Blue-Green 원두는", round(per_blue_green, 2), "% 입니다.")
```

[문제2] 커피원두 대륙별 + 색상별로 산도가 얼마나 다른가요? 산도를 확인할 수 있는 그래프를 그려주세요. (data['Acidity]>6)
- data_acidity = data_*[data_*['Acidity']>6] (data_*: 특정대륙데이터)
- plt.hist(data_acidity['Acidity'], bins=100, edgecolor='black')

```python
# 대륙 별 산도
africa_acidity = africa[africa['Acidity'] > 6]
north_america_acidity = north_america[north_america['Acidity'] > 6]
south_america_acidity = south_america[south_america['Acidity'] > 6]
asia_acidity = asia[asia['Acidity'] > 6]
oceania_acidity = oceania[oceania['Acidity'] > 6]


plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.hist(africa_acidity['Acidity'], bins=100, edgecolor='black')
plt.title('Africa')

plt.subplot(2, 3, 2)
plt.hist(north_america_acidity['Acidity'], bins=100, edgecolor='blue')
plt.title('North_america')

plt.subplot(2, 3, 3)
plt.hist(south_america_acidity['Acidity'], bins=100, edgecolor='red')
plt.title('South_america')

plt.subplot(2, 3, 4)
plt.hist(asia_acidity['Acidity'], bins=100, edgecolor='green')
plt.title('Asia')

plt.subplot(2, 3, 5)
plt.hist(oceania_acidity['Acidity'], bins=100, edgecolor='orange')
plt.title('Oceania')

plt.tight_layout()
plt.show()
```
