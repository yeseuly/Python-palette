```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../hw2/Diabetes Classification.csv")
```

[문제1] 
- 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
- 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)

```python
plt.subplot(2,1,1)
plt.scatter('Age','BMI', data=data)
plt.subplot(2,1,2)
plt.scatter('Age','Chol', data=data)
```

[문제2] 문제1의 그래프를 성별에 따라 분류하여 그래프로 나타내시오

```python
Male = data[data['Gender'] == 'M']
Female = data[data['Gender'] == 'F']

plt.subplot(2,2,1)
plt.scatter('Age','BMI', c = 'r', data=Female)
plt.subplot(2,2,2)
plt.scatter('Age','BMI', c = 'b', data=Male)

plt.subplot(2,2,3)
plt.scatter('Age','Chol', c = 'r', data=Female)
plt.subplot(2,2,4)
plt.scatter('Age','Chol', c = 'b', data=Male)
```

[문제3] 문제1의 그래프를 당뇨여부에 따라 분류하여 그래프로 나타내시오

```python
Diagnosis = data[data['Diagnosis'] == 1]
Nondiagnosis = data[data['Diagnosis'] == 0]

plt.subplot(2,2,1)
plt.scatter('Age','BMI', c = 'r', data=Diagnosis)
plt.subplot(2,2,2)
plt.scatter('Age','BMI', c = 'b', data=Nondiagnosis)
```

[문제4] 당뇨가 있는 사람과 없는 사람의 BMI, Chol 히스토그램을 하나의 그래프로 표현하세요

```python
Diagnosis = data[data['Diagnosis'] == 1]
Nondiagnosis = data[data['Diagnosis'] == 0]

plt.subplot(2,1,1)
plt.hist('BMI', data=Diagnosis)
plt.hist('BMI', data=Nondiagnosis)
plt.subplot(2,1,2)
plt.hist('Chol', data=Diagnosis)
plt.hist('Chol', data=Nondiagnosis)
```
