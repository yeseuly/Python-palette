```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../hw2/Diabetes Classification.csv")
```

[문제1] 
- 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
- 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)

```python
plt.figure()
plt.scatter(data['Age'], data['BMI'])
plt.xlabel('Age')
plt.ylabel('BMI')

plt.figure()
plt.scatter(data['Age'], data['Chol'])
plt.xlabel('Age')
plt.ylabel('Chol')

```

[문제2] 문제1의 그래프를 성별에 따라 분류하여 그래프로 나타내시오

```python
data_male = data[data['Gender']=='M']
data_female = data[data['Gender']=='F']
```

```python
plt.figure()
plt.scatter(data_male['Age'], data_male['BMI'],c='b',alpha=0.2,label='Male')
plt.scatter(data_female['Age'], data_female['BMI'],c='r',alpha=0.2,label='Female')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(loc='upper right')

plt.figure()
plt.scatter(data_male['Age'], data_male['Chol'],c='g',alpha=0.2,label='Male')
plt.scatter(data_female['Age'], data_female['Chol'],c='orange',alpha=0.2,label='Female')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend(loc='upper right')
```

[문제3] 문제1의 그래프를 당뇨여부에 따라 분류하여 그래프로 나타내시오

```python
data_Diagnosis1 = data[data["Diagnosis"]==1] #당뇨있
data_Diagnosis0 = data[data["Diagnosis"]==0] #당뇨없
```

```python
plt.figure(figsize=(10,8))


plt.subplot(2,1,1)
plt.scatter(data_Diagnosis1['Age'],data_Diagnosis1['BMI'],alpha=0.2,label='O',color='orange')
plt.scatter(data_Diagnosis0['Age'],data_Diagnosis0['BMI'],alpha=0.2,label='X')
plt.ylabel('BMI')
plt.legend(loc='upper right')

plt.subplot(2,1,2)
plt.scatter(data_Diagnosis1['Age'],data_Diagnosis1['Chol'],alpha=0.2,label='O',color='orange')
plt.scatter(data_Diagnosis0['Age'],data_Diagnosis0['Chol'],alpha=0.2,label='X')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend(loc='upper right')
```
