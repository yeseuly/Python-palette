```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../hw2/Diabetes Classification.csv")
```

[문제1] 
- 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
- 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)

```python
age = df['Age']
chol = df['Chol']
bmi = df['BMI']

plt.figure(figsize=(15, 6))

plt.scatter(age, chol, color='blue', alpha=0.5, label='Chol')
plt.scatter(age, bmi, color='red', alpha=0.5, label='BMI')

plt.title('Age - Chol / BMI')
plt.xlabel('Age')
plt.ylabel('Chol / BMI')

plt.grid(True)
plt.legend()
plt.show()
```

[문제2] 문제1의 그래프를 성별에 따라 분류하여 그래프로 나타내시오

```python
male = df[df['Gender'] == 'M']
female = df[df['Gender'] == 'F']

age_male = male['Age']
chol_male = male['Chol']
bmi_male = male['BMI']

age_female = female['Age']
chol_female = female['Chol']
bmi_female = female['BMI']

plt.figure(figsize=(15, 6))

plt.scatter(age_male, chol_male, color='blue', alpha=0.5, label='Male Chol')
plt.scatter(age_female, chol_female, color='red', alpha=0.5, label='Female Chol')

plt.scatter(age_male, bmi_male, color='green', alpha=0.5, label='Male BMI')
plt.scatter(age_female, bmi_female, color='orange', alpha=0.5, label='Female BMI')

plt.title('Gender')
plt.xlabel('Age')
plt.ylabel('Chol / BMI')

plt.grid(True)
plt.legend()
plt.show()
```

[문제3] 문제1의 그래프를 당뇨여부에 따라 분류하여 그래프로 나타내시오

```python
diag = df[df['Diagnosis'] == 1]
nondiag = df[df['Diagnosis'] == 0]

age_diag = diag['Age']
chol_diag = diag['Chol']
bmi_diag = diag['BMI']

age_nondiag = nondiag['Age']
chol_nondiag = nondiag['Chol']
bmi_nondiag = nondiag['BMI']


plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.scatter(age_diag, chol_diag, color='blue', alpha=0.5, label='Diag Chol')
plt.scatter(age_nondiag, chol_nondiag, color='red', alpha=0.5, label='Nondiag Chol')

plt.title('Diag')
plt.xlabel('Age')
plt.ylabel('Chol')

plt.grid(True)
plt.legend()



plt.subplot(2, 1, 2)
plt.scatter(age_diag, bmi_diag, color='green', alpha=0.5, label='Diag BMI')
plt.scatter(age_nondiag, bmi_nondiag, color='orange', alpha=0.5, label='Nondiag BMI')

plt.title('nondiag')
plt.xlabel('Age')
plt.ylabel('BMI')

plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
