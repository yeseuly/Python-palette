```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Diabetes Classification.csv")
```

```python
age = df['Age']
chol = df['Chol']
bmi = df['BMI']

# 그래프 크기 지정
plt.figure(figsize=(10, 6))

# 나이에 대한 Chol의 산점도 그래프 그리기
plt.scatter(age, chol, color='blue', alpha=0.5, label='Chol')

# 나이에 대한 BMI의 산점도 그래프 그리기
plt.scatter(age, bmi, color='green', alpha=0.5, label='BMI')

# 그래프 제목 설정
plt.title('나이에 대한 Chol과 BMI의 산점도 그래프')

# x축 레이블 설정
plt.xlabel('나이')

# y축 레이블 설정 (Chol과 BMI를 함께 그리므로 레이블이 중복되지 않도록 함)
plt.ylabel('Chol / BMI')

# 그리드 표시
plt.grid(True)

# 범례 표시
plt.legend()

# 그래프 출력
plt.show()
```

[문제1] 
- 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
- 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)

```python
# 남성과 여성을 나눔
male = df[df['Gender'] == 'M']
female = df[df['Gender'] == 'F']

# '나이', 'Chol', 'BMI' 열을 선택
age_male = male['Age']
chol_male = male['Chol']
bmi_male = male['BMI']

age_female = female['Age']
chol_female = female['Chol']
bmi_female = female['BMI']

# 그래프 크기 지정
plt.figure(figsize=(15, 6))

# 남성에 대한 Chol의 산점도 그래프 그리기
plt.scatter(age_male, chol_male, color='blue', alpha=0.5, label='Male Chol')

# 여성에 대한 Chol의 산점도 그래프 그리기
plt.scatter(age_female, chol_female, color='red', alpha=0.5, label='Female Chol')

# 남성에 대한 BMI의 산점도 그래프 그리기
plt.scatter(age_male, bmi_male, color='green', alpha=0.5, label='Male BMI')

# 여성에 대한 BMI의 산점도 그래프 그리기
plt.scatter(age_female, bmi_female, color='orange', alpha=0.5, label='Female BMI')

# 그래프 제목 설정
plt.title('Sex')

# x축 레이블 설정
plt.xlabel('나이')

# y축 레이블 설정 (Chol과 BMI를 함께 그리므로 레이블이 중복되지 않도록 함)
plt.ylabel('Chol / BMI')

# 그리드 표시
plt.grid(True)

# 범례 표시
plt.legend()

# 그래프 출력
plt.show()
```

[문제2] 문제1의 그래프를 성별에 따라 분류하여 그래프로 나타내시오

[문제3] 문제1의 그래프를 당뇨여부에 따라 분류하여 그래프로 나타내시오

```python
# 당뇨 여부를 나눔
diag = df[df['Diagnosis'] == 1]
nondiag = df[df['Diagnosis'] == 0]

# '나이', 'Chol', 'BMI' 열을 선택
age_diag = diag['Age']
chol_diag = diag['Chol']
bmi_diag = diag['BMI']

age_nondiag = nondiag['Age']
chol_nondiag = nondiag['Chol']
bmi_nondiag = nondiag['BMI']

# 그래프 크기 지정
plt.figure(figsize=(15, 6))

# Diag에 대한 Chol의 산점도 그래프 그리기
plt.scatter(age_diag, chol_diag, color='blue', alpha=0.5, label='Diag Chol')

# Nondiag에 대한 Chol의 산점도 그래프 그리기
plt.scatter(age_nondiag, chol_nondiag, color='red', alpha=0.5, label='Nondiag Chol')

# Diag에 대한 BMI의 산점도 그래프 그리기
plt.scatter(age_diag, bmi_diag, color='green', alpha=0.5, label='Diag BMI')

# Nondiag에 대한 BMI의 산점도 그래프 그리기
plt.scatter(age_nondiag, bmi_nondiag, color='orange', alpha=0.5, label='Nondiag BMI')

# 그래프 제목 설정
plt.title('Diag')

# x축 레이블 설정
plt.xlabel('나이')

# y축 레이블 설정 (Chol과 BMI를 함께 그리므로 레이블이 중복되지 않도록 함)
plt.ylabel('Chol / BMI')

# 그리드 표시
plt.grid(True)

# 범례 표시
plt.legend()

# 그래프 출력
plt.show()
```

[문제4] 당뇨가 있는 사람과 없는 사람의 BMI, Chol 히스토그램을 하나의 그래프로 표현하세요

```python

# 남성과 여성을 나눔
male = df[df['Gender'] == 'M']
female = df[df['Gender'] == 'F']

# 그래프 크기 지정
plt.figure(figsize=(15, 6))

# 남성에 대한 그래프
plt.subplot(1, 2, 1)  # 1행 2열 중 첫 번째 서브플롯 선택
plt.scatter(male['Age'], male['Chol'], color='blue', alpha=0.5, label='Male Chol')
plt.scatter(male['Age'], male['BMI'], color='green', alpha=0.5, label='Male BMI')
plt.title('Male')  # 서브플롯 제목
plt.xlabel('Age')
plt.ylabel('Chol / BMI')
plt.grid(True)
plt.legend()

# 여성에 대한 그래프
plt.subplot(1, 2, 2)  # 1행 2열 중 두 번째 서브플롯 선택
plt.scatter(female['Age'], female['Chol'], color='red', alpha=0.5, label='Female Chol')
plt.scatter(female['Age'], female['BMI'], color='orange', alpha=0.5, label='Female BMI')
plt.title('Female')  # 서브플롯 제목
plt.xlabel('Age')
plt.ylabel('Chol / BMI')
plt.grid(True)
plt.legend()

# 그래프 출력
plt.tight_layout()  # 서브플롯 간의 간격 조정
plt.show()
```
