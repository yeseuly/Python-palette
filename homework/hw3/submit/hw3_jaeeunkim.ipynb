[문제1] 
- 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
- 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)

```python
plt.figure(figsize=(7,7))
plt.subplot(2, 1, 1)
plt.scatter('Age', 'BMI', data=data)
plt.subplot(2, 1, 2)
plt.scatter('Age', 'Chol', data=data)
```

[문제2] 문제1의 그래프를 성별에 따라 분류하여 그래프로 나타내시오

```python
data_female = data[data['Gender'] == 'F']
data_male = data[data['Gender'] == 'M']

plt.figure(figsize=(7,7))
plt.subplot(2, 1, 1)
plt.scatter('Age', 'BMI', data=data_female)
plt.subplot(2, 1, 2)
plt.scatter('Age', 'BMI', data=data_male)

plt.legend()
```

[문제3] 문제1의 그래프를 당뇨여부에 따라 분류하여 그래프로 나타내시오

```python
data_no = data[data['Diagnosis'] == 0]
data_yes = data[data['Diagnosis'] == 1]

plt.figure(figsize=(7,7))
plt.subplot(2, 1, 1)
plt.scatter('Age', 'BMI', data=data_no)
plt.subplot(2, 1, 2)
plt.scatter('Age', 'BMI', data=data_yes)
```

[문제4] 당뇨가 있는 사람과 없는 사람의 BMI, Chol 히스토그램을 하나의 그래프로 표현하세요

```python
data_no = data[data['Diagnosis'] == 0]
data_yes = data[data['Diagnosis'] == 1]

plt.subplot(2,1,1)
plt.hist('BMI', data=data_no)
plt.hist('BMI', data=data_yes)
plt.subplot(2,1,2)
plt.hist('Chol', data=data_no)
plt.hist('Chol', data=data_yes)
```
