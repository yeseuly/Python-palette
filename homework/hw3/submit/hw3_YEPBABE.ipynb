```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../hw2/Diabetes Classification.csv")
```

[문제1] 
- 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
- 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)

```python
plt.figure()
plt.scatter(data['Age'],data['BMI'],c='orange')
plt.xlabel('Age')
plt.ylabel('BMI')

plt.figure()
plt.scatter(data['Age'],data['Chol'])
plt.xlabel('Age')
plt.ylabel('Chol')
```

```python
data_male = data[data['Gender']=="M"] #data에서 남자 Male만 추출한 것
data_female = data[data['Gender']=="F"]
```

[문제2] 문제1의 그래프를 성별에 따라 분류하여 그래프로 나타내시오

```python

plt.figure(figsize=(8,7))
plt.subplot(2,2,1)
plt.scatter(data_male['Age'],data_male['BMI'],c='orange')
plt.ylabel('BMI')
plt.title("Male")

plt.subplot(2,2,2)
plt.scatter(data_female['Age'],data_female['BMI'],c='b')
plt.ylabel('BMI')
plt.title("Female")

plt.subplot(2,2,3)
plt.scatter(data_male['Age'],data_male['Chol'],c='r')
plt.xlabel('Age')
plt.ylabel('Chol')

plt.subplot(2,2,4)
plt.scatter(data_female['Age'],data_female['Chol'])
plt.xlabel('Age')
plt.ylabel('Chol')
```

[문제3] 문제1의 그래프를 당뇨여부에 따라 분류하여 그래프로 나타내시오

```python
data_Diagnosis1 = data[data["Diagnosis"]==1]
data_Diagnosis0 = data[data["Diagnosis"]==0]

######################
#  [문제1] 
#  - 나이에 따른 BMI 수치를 Scattering Graph로 표현하시오 (X=Age, Y=BMI)
#  - 나이에 따른 Chol 수치를 Scattering Graph로 표현하시오 (X=Age, Y=Chol)
#####################
```

```python
plt.figure(figsize=(17,8))
plt.subplot(2,1,1)
plt.scatter(data_Diagnosis1['Age'],data_Diagnosis1['BMI'],c='orange', label='Diagnosis YES')
plt.scatter(data_Diagnosis0['Age'],data_Diagnosis0['BMI'],c='skyblue', label='Diagnosis NO')
plt.ylabel('BMI')
plt.legend(loc='upper right')


plt.subplot(2,1,2)
plt.scatter(data_Diagnosis1['Age'],data_Diagnosis1['Chol'],c='orange',label='Diagnosis YES')
plt.scatter(data_Diagnosis0['Age'],data_Diagnosis0['Chol'],c='skyblue',label='Diagnosis NO')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend(loc='upper right')
```

[문제4] 당뇨가 있는 사람과 없는 사람의 BMI, Chol 히스토그램을 하나의 그래프로 표현하세요

```python
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
a1 = plt.hist(data_Diagnosis1['BMI'], density=False, bins=30, color='r', alpha=0.7, edgecolor='black', label='Diagnosis YES')
a0 = plt.hist(data_Diagnosis0['BMI'], density=False, bins=30, color='b', alpha=0.7, edgecolor='black', label='Diagnosis NO')
plt.xlabel('BMI')
plt.ylabel('How many PEOPLE')
plt.legend()

plt.subplot(1,2,2)
b1 = plt.hist(data_Diagnosis1['Chol'], density=False, bins=30, color='r', alpha=0.7, edgecolor='black', label='Diagnosis YES')
b0 = plt.hist(data_Diagnosis0['Chol'], density=False, bins=30, color='b', alpha=0.7, edgecolor='black', label='Diagnosis NO')
plt.xlabel('Chol')
plt.legend()
```
