[문제2] 당뇨병에 걸린 사람(Diagnosis==1)의 남녀 비율은 어떻게 되나요? 평균 연령은 어떻게 되나요?
- 당뇨병에 걸리지 않은 사람들은 총 몇명인가요?
- 당뇨병에 걸리지 않은 남녀 비율은 어떻게 되나요?

```python


D_d_people = data[data['Diagnosis']==1]
D_nd_people = data[data['Diagnosis']==0]

Sum_diag_people = len(D_d_people)
print("당뇨에 걸린이들 숫자 " + str(Sum_diag_people))

Sum_nodiag_people = len(D_nd_people)
print("당뇨에 안 걸린이들 숫자 " + str(Sum_nodiag_people))

Mean_D_p_Age = np.mean(D_d_people['Age'])
print("당뇨 걸린이들 평균 연령 " + str(Mean_D_p_Age))

D_d_man = data[(data['Diagnosis']==1) & (data['Gender']=="M")]

D_d_woman = data[(data['Diagnosis']==1) & (data['Gender']=="F")]

Sum_diag_man = len(D_d_man)
print("당뇨 남자 숫자 " + str(Sum_diag_man))

Sum_diag_woman = len(D_d_woman)
print("당뇨 여자 숫자 " + str(Sum_diag_woman))

print("당뇨 남자 비율 " + str(Sum_diag_man / Sum_diag_people * 100) + "%")
print("당뇨 여자 비율 " + str(Sum_diag_woman / Sum_diag_people * 100) + "%")

D_nd_man = data[(data['Diagnosis']==0) & (data['Gender']=="M")]
Sum_nd_man = len(D_nd_man)
D_nd_woman = data[(data['Diagnosis']==0) & (data['Gender']=="F")]
Sum_nd_woman = len(D_nd_woman)

print("안당뇨 남자 비율 " + str(Sum_nd_man / Sum_nodiag_people * 100) + "%")
print("안당뇨 여자 비율 " + str(Sum_nd_woman / Sum_nodiag_people * 100) + "%")




Mean_D_M_Age = np.mean(D_d_man['Age'])
Mean_D_F_Age = np.mean(D_d_woman['Age'])
print("당뇨 남자 평균 연령 " + str(Mean_D_M_Age))
print("당뇨 여자 평균 연령 " + str(Mean_D_F_Age))

```
