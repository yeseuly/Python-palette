- for 반복문: num_list의 원소를 반복해서 출력한다

```
num_list = [1, 2, 3]
for num in num_list:
    print(num)
```

- for 반복문: 리스트의 원소를 반복해서 2의 배수와 함께 출력한다.

```
for num in [1, 2, 3]:
    print("기존: ", num)
    print("2배: ", num * 2)
    print("\n")            
```

- for 반복문: 리스트의 원소를 반복해서 2의 배수를 계산하고, 빈 리스트(double)에 2의 배수를 원소로 추가한다

```
double = []
for num in [1, 2, 3]:
    double.append(num * 2)

print(double)            
```
