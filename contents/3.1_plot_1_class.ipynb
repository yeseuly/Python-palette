# Matplotlib 기본 사용법
### Pyplot 소개

matplotlib.pyplot MATLAB과 비슷하게 명령어 스타일로 동작하는 파이썬 시각화 모듈입니다.

### 기본 그래프

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [1, 2, 3, 4]

plt.plot(x, y)
plt.xticks(x)
plt.yticks(y)
```

### 스타일 지정

: x, y 값 인자에 대해 그래프 선의 색상과 형태를 지정하는 포맷 문자열을 세번째 인자에 입력할 수 있습니다.

```python
x=[1, 2, 3, 4] 
y=[1, 4, 9, 16]

plt.plot(x, y, color='red', linewidth=3, linestyle='--')
```

```python
x=[1, 2, 3, 4] 
y=[1, 4, 9, 16]

plt.plot(x, y, 'ro--')
```

- 첫번째값은 색상, 두번째값은 마커 스타일, 세번째값은 선 스타일을 의미합니다.
- 기본 포맷 문자열은 ‘b-‘이고 파란색(‘blue’)와 선(‘-‘)을 의미합니다.
- ‘rs-’는 빨간색(‘red’)와 사각형(‘ㅁ’), 선 모양('--')을 의미합니다.

### 색상
: 색상을 지정하는 방법은 색 이름이나 약자 또는 # 문자로 시작되는 RGB코드를 사용합니다. 색상코드는 다음 사이트에서 참고합니다.

### 스타일 지정 예시 (pdf 파일 참고)

- color='green'
- linewidth=3
- linestyle='solid'
- markersize=15
- markeredgecolor='orange'
- markeredgewidth=5

```python
x = [10, 20, 30, 40] 
y = [1, 5, 10, 15]

plt.plot(x, y, color='green', linewidth=3, linestyle='solid', marker="o", markersize=15, 
         markeredgecolor='orange', markeredgewidth=5)
```

### 그래프 범위 설정

: 그래프를 그릴때 원하는 범위를 설정해줍니다.

```python
# xlim = [0, 50] ylim = [-30, 30]

x = [10, 20, 30, 40] 
y = [1, 5, 10, 15]

plt.plot(x, y, color='green', linewidth=3, linestyle='solid', marker="o", markersize=15, 
         markeredgecolor='orange', markeredgewidth=5)
plt.xlim([0, 50])
plt.ylim([-30, 30])
```

### 그래프 tick 설정

: tick 옵션으로 x축과 y축의 라벨을 설정할 수 있습니다.

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

# xticks = [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]
# yticks = [-1, 0, 1]

plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, 1])
```

다음과 같이 Latex 문자식으로 표현도 가능합니다.

```python
# xticks-latex = [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
# yticks-latex = ["Low", "Zero", "High"]

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)

plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
```

### 여러 개의 그래프 그리기

```python
t = np.arange(0., 5., 0.2)

# 'r--': t 'b*': t**2 'gs':t**3

plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'b*')
plt.plot(t, t**3, 'gs')

# plt.plot(t, t, 'r--', t, t**2, 'b*', t, t**3, 'gs')
```

### 범례

: 여러 그래프를 동시에 그릴 경우 각각의 선이 무엇을 나타내는지 표시할 필요가 있습니다. legend 함수를 통해 표현할 수 있으며, loc옵션을 통해 위치를 설정할수도 있습니다.

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, linestyle='dashed', linewidth=2, label='cosine', color='magenta')
plt.plot(X, S, linestyle='dashed', linewidth=1, label='sine', color='lightseagreen')
plt.legend(loc='upper left')
```

### x축, y축, 제목 설정

: x축, y축의 제목을 설정하고, 가장 큰 제목인 title을 설정합니다.

- title: Cosine Plot
- xlabel: time
- ylabel: amplitude

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)

plt.title("Cosine Plot")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

### 그래프의 구조
- matplotlib는 Figure 객체, Axes 객체, Axis 객체 등으로 구성되어 있습니다.
- Figure는 그림이 그려지는 캔버스나 종이, Axes는 하나의 그래프, Axis는 가로축이나 세로축 등의 축을 의미합니다.
- Figure는 주로 여러개의 그래프를 동시에 그리고 싶을때, 그래프의 크기를 설정하고 싶을때 사용합니다.

### figure 사이즈 설정
: (10,2) 사이즈로 그래프를 설정합니다.

```python
np.random.seed(0)

plt.figure(figsize=(10,2))
plt.plot(np.random.randn(100))
```

### subplot

- 하나의 윈도우(figure)안에 여러개의 그래프를 표현하고 싶을때 사용하는 명령어입니다.
- 그리드(grid) 형태의 Axes 객체들을 생성하는데 Figure가 행렬(matrix)이고 Axes가 행렬의 원소라고 생각하면 됩니다.
- subplot은 3개의 인자를 가지는데 처음 두개 인자는 전체 그리드 행렬의 모양을 의미하고, 세번째 인자는 여러개의 그래프에서 어느것인지를 의미하는 숫자입니다.

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.figure(figsize=(7, 4))
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.xlabel("time")
plt.ylabel("Damped Oscillation")
plt.title("A tale of 2 subplot")

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r^-')
plt.xlabel("time")
plt.ylabel("Undampled")
```

### axes
subplots 명령어를 사용해서 복수의 Axes 객체를 동시에 생성할 수도 있습니다다.

```python
plt.figure(figsize=(7, 6))
plt.subplot(2, 2, 1)
plt.plot(np.random.randn(5))
plt.title("Axes 1")
plt.subplot(2, 2, 2)
plt.plot(np.random.randn(5))
plt.title("Axes 2")
plt.subplot(2, 2, 3)
plt.plot(np.random.randn(5))
plt.title("Axes 3")
plt.subplot(2, 2, 4)
plt.plot(np.random.randn(5))
plt.title("Axes 4")
```
