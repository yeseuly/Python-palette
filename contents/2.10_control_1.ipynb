- if 조건문 (조건식: a 변수의 값을 2로 나눈 나머지가 0과 같다)

```python
a = 13

if a % 5 == 0:
    print("a는 5의 배수")   
elif a % 5 == 1:
    print("나머지 1")
elif a % 5 == 2:
    print("나머지 2")
else:
    print("...")  
```

- if 조건문 (조건식: a 변수의 값을 2로 나눈 나머지가 0이 아니다)

```python
if a % 2 != 0:
    print("a는 홀수")
```

- if ~ else 조건문 (조건식: b 변수의 값을 2로 나눈 나머지가 0과 같다)

```python
b = 7

if b % 2==0:
    print("b는 짝수")
else:
    print("b는 홀수")
```

- if 조건식: c 변수의 값이 0보다 크다.
- elif 조건식: c 변수의 값이 0보다 작다.
- else 조건식: 없음 (c 변수의 값이 0이다)


```python
c = -1

if c > 0:
    print("c는 양수")
elif c < 0:
    print("c는 음수")  
else:
    print("c는 0")
```
